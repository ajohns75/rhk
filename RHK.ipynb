# Data Preparation

```python
import pandas as pd
import numpy as np
from datetime import timedelta, datetime, date
```

```python
kid_df = pd.read_excel('rhk_kids.xlsx', skiprows=1)
```

```python
kid_df.rename(columns = {'JOIN DATE': 'join_date',
                         'Unnamed: 1': 'photo_release',
                        'FIRST': 'first_name',
                        'LAST': 'last_name',
                        'ADDRESS': 'address',
                        'DOB': 'birth_month',
                        'Unnamed: 9':'birth_day',
                        'Unnamed: 10': 'birth_year',
                        'NOTES':'notes',
                        'SEX':'gender'}, inplace=True)
```

```python
# Remove divider row 
kid_df.drop(index=[275, 276, 277, 278], inplace=True)
```

```python
#############
# Cities, states, and zip codes
#############
city_df = kid_df['CITY, ST  ZIP'].str.split(' ', expand=True)

# clean up cities
city_df['city'] = np.where((city_df[1].str.len() > 2), city_df[0] + ' ' + city_df[1], city_df[0])
clean_cities = city_df.city.str.split(',', expand=True)
clean_cities[0].unique()
clean_cities['city'] = np.where(clean_cities[0].str.startswith("R"), 'Rockford', clean_cities[0])
clean_cities['city'].replace('City', np.NaN, inplace=True)
clean_cities['city'].replace(',', '', inplace=True)

# clean states 
city_df['state'] = np.where(city_df[1].str.contains('il', case=False), 'IL', np.NaN)
city_df['state'] = np.where((city_df['state'].isna()) & (city_df[2].str.contains('il', case=False)), 'IL', city_df['state'])
city_df['state'] = 'IL'

# clean zip codes
city_df[1].str.isdigit().sum()
city_df['zip_codes'] = np.where(city_df[2].str.isdigit() == True, city_df[2], np.NaN)
city_df['zip_codes'] = np.where(city_df[3].str.isdigit() == True, city_df[3], city_df['zip_codes'])
```

```python
city_df.state.unique()
```

```python
##############
# Primary parents name and phone numbers
##############
primary_parent_df = kid_df['PARENT/GUARDIAN & PHONE'].str.split("(\d)", n=1, expand=True)
primary_parent_df[['first_name', 'last_name']] = primary_parent_df[0].str.split(' ', n=1, expand=True)
primary_parent_df['phone_number'] = primary_parent_df[1]+primary_parent_df[2]
primary_parent_df[['last_name', 'relation']] = primary_parent_df['last_name'].str.split('\(', expand=True)
primary_parent_df = primary_parent_df[['first_name', 'last_name', 'relation', 'phone_number']]
primary_parent_df.relation = primary_parent_df.relation.replace(to_replace = ["\)"], regex=True, value="")

### Goofy data 
goofy_parent_data = primary_parent_df.loc[(primary_parent_df['phone_number'].str.len() > 12)]

goofy_parent_data['true_phone_number'] = goofy_parent_data['phone_number'].str.slice(0,12)
goofy_parent_data['remainders'] = goofy_parent_data['phone_number'].str.slice(12,)

# get extra parent numbers
goofy_numbers = goofy_parent_data['remainders'].str.split("(\d)", n=1, expand=True)
goofy_numbers['true_phone_number2'] = goofy_numbers[1] + goofy_numbers[2]

goofy_numbers[0] = goofy_numbers[0].replace(to_replace = ["or", 'OR', '\/', '\*', '\:', '\-'], regex=True, value="")

goofy_numbers['contact2'] = np.where(goofy_numbers['true_phone_number2'].str.contains('\('), 
                                    goofy_numbers['true_phone_number2'].str.slice(12,),
                                    goofy_numbers[0])
goofy_numbers['true_phone_number2'] = goofy_numbers['true_phone_number2'].str.slice(0,12)

goofy_data = pd.merge(goofy_parent_data, goofy_numbers, left_index=True, right_index=True)

goofy_data = goofy_data[['true_phone_number', 'contact2', 'true_phone_number2']]

primary_parents_clean = pd.concat([primary_parent_df, goofy_data], axis=1)
primary_parents_clean['phone_number'] = np.where(primary_parents_clean['true_phone_number'].isna()==False,
                                                 primary_parents_clean['true_phone_number'], primary_parents_clean['phone_number'])
```

```python
##############
# Alternate parents name and phone numbers
##############
alternate_parent_df = kid_df['ALTERNATE PICKUP/PHONE'].str.split("(\d)", n=1, expand=True)
alternate_parent_df[['first_name', 'last_name']] = alternate_parent_df[0].str.split(' ', n=1, expand=True)
alternate_parent_df['phone_number'] = alternate_parent_df[1]+alternate_parent_df[2]
alternate_parent_df[['last_name', 'relation']] = alternate_parent_df['last_name'].str.split('\(', expand=True)
alternate_parent_df = alternate_parent_df[['first_name', 'last_name', 'relation', 'phone_number']]
alternate_parent_df.relation = alternate_parent_df.relation.replace(to_replace = ["\)"], regex=True, value="")

### Goofy data 
goofy_alt_data = alternate_parent_df.loc[(alternate_parent_df['phone_number'].str.len() > 12)]

goofy_alt_data['true_phone_number'] = goofy_alt_data['phone_number'].str.slice(0,12)
goofy_alt_data['remainders'] = goofy_alt_data['phone_number'].str.slice(12,)

# get extra parent numbers
goofy_alt_numbers = goofy_alt_data['remainders'].str.split("(\d)", n=1, expand=True)
goofy_alt_numbers['true_phone_number2'] = goofy_alt_numbers[1] + goofy_alt_numbers[2]
goofy_alt_numbers[['extra', 'true_phone_number3']] = goofy_alt_numbers['true_phone_number2'].str.split(',', expand=True)

goofy_alt_numbers[0] = goofy_alt_numbers[0].replace(to_replace = ["or", 'OR', '\/', '\*', '\:', '\-', ',', '\&', '\?'], regex=True, value="")

goofy_alt_numbers['contact2'] = np.where(goofy_alt_numbers['true_phone_number2'].str.contains('\('), 
                                     goofy_alt_numbers['true_phone_number2'].str.slice(12,),
                                     goofy_alt_numbers[0])
goofy_alt_numbers['true_phone_number2'] = goofy_alt_numbers['true_phone_number2'].str.slice(0,12)

goofy_alt_data = pd.merge(goofy_alt_data, goofy_alt_numbers, left_index=True, right_index=True)
goofy_alt_data = goofy_alt_data[['true_phone_number', 'contact2', 'true_phone_number2', 'true_phone_number3']]

alternate_parents_clean = pd.concat([alternate_parent_df, goofy_alt_data], axis=1)
alternate_parents_clean['phone_number'] = np.where(alternate_parents_clean['true_phone_number'].isna()==False,
                                                  alternate_parents_clean['true_phone_number'], alternate_parents_clean['phone_number'])
```

```python
########### 
# Combine clean data
###########
# City, state, zip codes
kid_df['city'] = clean_cities['city']
kid_df['state'] = city_df['state']
kid_df['zip_code'] = city_df['zip_codes']

# Primary contact info
kid_df['primary_guardian_first_name'] = primary_parents_clean['first_name']
kid_df['primary_guardian_last_name'] = primary_parents_clean['last_name']
kid_df['primary_guardian_relation'] = primary_parents_clean['relation']
kid_df['primary_guardian_phone'] = primary_parents_clean['phone_number']

# Extra primary contact info
kid_df['primary_guardian2_title'] = primary_parents_clean['contact2']
kid_df['primary_guardian2_phone'] = primary_parents_clean['true_phone_number2']

# Alternate contact info
kid_df['alternate_guardian_first_name'] = alternate_parents_clean['first_name']
kid_df['alternate_guardian_last_name'] = alternate_parents_clean['last_name']
kid_df['alternate_guardian_relation'] = alternate_parents_clean['relation']
kid_df['alternate_guardian_phone'] = alternate_parents_clean['phone_number']

# Extra alternate contact info
kid_df['alternate_guardian2_title'] = alternate_parents_clean['contact2']
kid_df['alternate_guardian2_phone'] = alternate_parents_clean['true_phone_number2']
kid_df['alternate_guardian3_phone'] = alternate_parents_clean['true_phone_number3']

# Remove extra columns 
kid_df = kid_df.drop(['CITY, ST  ZIP', 'PARENT/GUARDIAN & PHONE', 'ALTERNATE PICKUP/PHONE'], axis=1)
```

```python
###########
# Add information / extra cleaning
###########
# Add teen / younger column based on index 
kid_df['age_group'] = np.NaN
kid_df['age_group'][:275] = 'young'
kid_df['age_group'][275:]= 'teen'

# Add full birthdays
kid_df['birth_year'] = np.where(kid_df['birth_year'] < 50, kid_df['birth_year']+2000, kid_df['birth_year'])
kid_df['full_birthday'] = kid_df[['birth_month', 'birth_day', 'birth_year']].apply(lambda x: '-'.join(x.values.astype(str)), axis="columns")
kid_df['full_birthday'] = np.where(kid_df['full_birthday'].str.contains('nan'), '01-01-1900', kid_df['full_birthday'])
kid_df['full_birthday'] = kid_df['full_birthday'].astype(str)


# Add age
def age(born):
    born = datetime.strptime(born, "%m-%d-%Y").date()
    today = date.today()
    return today.year - born.year - ((today.month, 
                                      today.day) < (born.month, 
                                                    born.day))
kid_df['age'] = kid_df['full_birthday'].apply(age)
kid_df['full_birthday'].loc[(kid_df['full_birthday'] == '01-01-1900')] = np.NaN
kid_df['age'].loc[(kid_df['age'] > 30)] = np.NaN

# Full addresses to generate lat / long
address_df = pd.DataFrame()
address_df['add_no_hash'] = kid_df["address"].str.split("#", expand=True)[0]
address_df['add_no_apt'] = kid_df["address"].str.split("Apt", expand=True)[0]
address_df['address_no_apt'] = np.where(address_df['add_no_apt'].str.contains('\#'), address_df['add_no_hash'], 
                                       address_df['add_no_apt'])
kid_df['address_no_apt'] = address_df['address_no_apt']

kid_df['full_address'] = kid_df['address'] + ', ' + kid_df['city']+ ', '+kid_df['state']
kid_df['full_address_no_apt'] = kid_df['address_no_apt'] + ', ' + kid_df['city']+ ', '+kid_df['state']

# Add count column to combine households later 
kid_df['number_of_kids'] = 1
```

```python
############
# Generate location data
############
import requests
import urllib.parse

location_df = kid_df[(kid_df['full_address'].isna() == False)]
kid_df['lat'] = np.NaN
kid_df['long'] = np.NaN
```

```python
houses = plot_house_df.dropna(subset=['lat', 'long'])
houses['number_of_kids'] = 1
```

```python
households = houses.groupby(['lat', 'long', 'full_address_no_apt'])['number_of_kids'].sum().to_frame().reset_index()
households[['lat', 'long', 'number_of_kids']] = households[['lat', 'long', 'number_of_kids']].astype(float)
households['n_kids'] = households['number_of_kids'].astype(str)
```

```python
for idx, row in households.iterrows():
    Marker([row['lat'], row['long']]).add_to(m1)
```

```python
# Rock House lat/long
url = 'https://nominatim.openstreetmap.org/search/' + urllib.parse.quote('1325 7th St., Rockford, IL') +'?format=json'
response = requests.get(url).json()

m3 = folium.Map(location=[response[0]['lat'], response[0]['lon']], tiles='cartodbpositron', zoom_start=13)
```

```python
mc = MarkerCluster()

for idx, row in households.iterrows():
    if not math.isnan(row['long']) and not math.isnan(row['lat']):
        mc.add_child(Marker([row['lat'], row['long']]))
m3.add_child(mc)
m3      
```

```python
m4 = folium.Map(location=[response[0]['lat'], response[0]['lon']], tiles='cartodbpositron', zoom_start=13)
mc = MarkerCluster()
for idx, row in households.iterrows():
    lat = households.at[idx, 'lat']
    long = households.at[idx, 'long']
        
    popup = 'Address: ' + str(households.at[idx, 'full_address_no_apt']) + '<br>' +'Number of Kids: ' + households.at[idx, 'n_kids']
    folium.Marker(location = [row['lat'], row['long']], popup=popup, icon=folium.Icon(color='blue', icon='home')).add_to(mc)

    m4.add_child(mc)
m4  
m3.save('rhk_households.html')
webbrowser.open('rhk_households.html', new=2)
```
